```python
from scipy.integrate import quad
def f(x):
  return x
area=quad(f,0,2)
print(area)
```

```python
from scipy.integrate import dblquad
def f(x,y):
  return x*y
area=dblquad(f,0,2,0,5)
print(area)
```
