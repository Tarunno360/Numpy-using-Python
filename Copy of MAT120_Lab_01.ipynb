```python
#another way to do the derivatives
x=symbols('x')
f=2*x**3+x**2+3*x+5
df_first=diff(f, x)
df_second=diff(f, x, 2)
print(df_first)
print(df_second)
print(f.subs(x, 4))
print(df_first.subs(x, 4))
```

```python
#multple variable er jonno
x, y, z=symbols('x y z')
expr=x**3+4*x*y-z
exprvalue=expr.subs([(x, 2), (y, 4), (z, 2)])
dfExpr=diff(expr, x)
df2Expr=diff(expr, x, 2)
print(exprvalue)
print(dfExpr)
print(df2Expr)
dfExpr_y=diff(expr, y)
df2Expr_y=diff(expr, y, 2)
print(dfExpr_y)
print(df2Expr_y)
```

```python
#y=f(x)=2x^3-6x^2-18x+7 maximapoint(-1, 17) minima(3, -47) minima -47 and maxima 17

from sympy import *

x = Symbol('x')
f_x=2*x**3-6*x**2-18*x+7
first_diff=diff(f_x, x)
# second_diff=diff(f_x, x, 2)
a=solve(first_diff, x)
for i in a:
    if i<0:
        maxima= f_x.subs(x, i)
        print(f'Maxima is {maxima} and maximapoint is {i, maxima}.')
    elif i>0:
        minima=f_x.subs(x, i)
        print(f'Minima is {minima} and minimapoint is {i, minima}.')
    else:
        print('In conclusive.')
```
